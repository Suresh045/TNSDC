```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
dataset=pd.read_csv("VGN.csv")
```

```python
dataset
```

```python
# Initialize LabelEncoder
label_enc = LabelEncoder()

# Encoding Designation
dataset["Designation"] = label_enc.fit_transform(dataset["Designation"])

# Encoding Area of Purchase
dataset["Area of Purchase"] = label_enc.fit_transform(dataset["Area of Purchase"])

dataset["Existing EMI"] = label_enc.fit_transform(dataset["Existing EMI"])
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
independent=dataset[['Designation', 'Area of Purchase', 'No of Purchase', 'Existing EMI']]
dependent=dataset[["Amount"]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.20,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression(fit_intercept=True, n_jobs=None)
regressor.fit(X_train,y_train)
```

```python
weight=regressor.coef_
weight
```

```python
bais=regressor.intercept_
bais
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename="Multiple LinearRegression_salary.sav"
pickle.dump(regressor,open(filename,'wb'))
```
